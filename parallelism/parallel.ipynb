```python
import time
import subprocess

batch_sizes = [1, 10, 50, 100, 500, 600, 700, 800, 900, 1000]

for batch in batch_sizes:
    print(f"\nTesting with batch size: {batch}")
    start = time.time()
    subprocess.run(["python", "sph_parallel_shared_memory_batched.py", str(batch)])
    elapsed = time.time() - start
    print(f"Batch size: {batch} - Total elapsed time: {elapsed:.4f} seconds")
```
